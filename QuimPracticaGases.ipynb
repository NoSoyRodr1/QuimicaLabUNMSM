```python
boyle = pd.read_csv("data (1).csv")
boyle
```

```python
plt.scatter(1/boyle["x"],boyle["y"])
```

```python
from sklearn.linear_model import LinearRegression

modelo = LinearRegression()

modelo.fit(boyle[['x']], boyle['y'])
```

```python
intercepto = modelo.intercept_
pendiente = modelo.coef_[0]
```

```python
ecuacion = f'x = {pendiente:.2f}y + {intercepto:.2f}'
print("La ecuación de la ley de Boyle será: ",ecuacion)
```
