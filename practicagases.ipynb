2. Extraer datos químicos de la web y hacer una exploración gráfica(*opcional aplicar un modelo)

```python
#Extraer datos quimicos de la web y hacer una exploracion grafica. aplicar un modelo
import pandas as pd

url = "https://en.wikipedia.org/wiki/Periodic_table"
df = pd.read_html(url, match="Element")[0]

# Limpiar los datos
df = df.dropna()
df["Density"] = df["Density"].str.replace(" g/cm³", "")
df["Density"] = df["Density"].astype(float)

# Trazar el gráfico
plt.plot(df["Atomic number"], df["Density"])
plt.xlabel("Atomic number")
plt.ylabel("Density (g/cm³)")
plt.show()

```
